Line Finder

```python
# Line finder
def line_user_find(file):
    switch_for_line = True
    line_no = 1
    while switch_for_line:
        uname = input("Enter username: ")
        for row in file:
            if row[1] == uname:
                print("User Found:", uname)
                uname_pwd = [row[0], row[1], row[2], row[3]]
                
                return uname_pwd,line_no
                switch_for_line = False
            else:
                line_no += 1

```

```python
def find():
    try:
        with open('accounts.txt', 'r') as file:
            file_reader = csv.reader(file)
            s = user_find(file_reader)
            if s == True:
                pass
            else:
                print("User not Found")
            file.close()
    except:
        print("No File Found!! Make a new account first")


def user_find(file):
    user = input("Enter username: ")
    for row in file:
        if row[1] == user:
            print("User Found:", user)
            user_found = [row[0], row[1], row[2], row[3]]
            print("\n\nName:", user_found[0])
            print("Username:", user_found[1])
            print('Password:', user_found[2])
            print('Bank Balance:', user_found[3])
            time.sleep(5)
            print("\n\n\n\n\n\n\n")
            return True
        else:
            pass
```

```python

def cash_deposit():
    try:
        with open('accounts.txt', 'r') as file:
            file_reader = csv.reader(file)
            depositor(file_reader)
            file.close()
    except:
        print("No File Found!! Make a new account first")


def depositor(file):
    user_data,lineno = line_user_find(file)
    pass_check(user_data)
    dep_amnt = int(input("Enter amount you want to deposit: "))
    prv_amnt = int(user_data[3])
    prv_amnt += dep_amnt

    deposit_here(lineno, user_data, prv_amnt)
    print("Cash Deposit Successful \n\n\n\n\n")
    

def deposit_here(lineNo, user_found, amnt):
    name = user_found[0] + ','
    uname = user_found[1] + ','
    pass1 = user_found[2] + ','
    pysa = str(amnt) + '\n'
    isSkipped = False
    index = 1
    with open("accounts.txt", 'r') as readObj, open("temp_file", 'w') as writeObj:
        for line in readObj:
            if index != lineNo:
                writeObj.write(line)
            else:
                isSkipped = True
            index += 1
        writeObj.write(name)
        writeObj.write(uname)
        writeObj.write(pass1)
        writeObj.write(pysa)
    if isSkipped:
        os.remove("accounts.txt")
        os.rename("temp_file", "accounts.txt")
    else:
        os.remove("temp_file")
```

```python

def cash_withdraw():
    try:
        with open('accounts.txt', 'r') as file:
            file_reader = csv.reader(file)
            withdrawer(file_reader)
            file.close()
    except:
        print("No File Found!! Make a new account first")


def withdrawer(file):
    user_data,lineno = line_user_find(file)
    pass_check(user_data)
    with_amnt = int(input("Enter amount you want to Withdraw: "))
    prv_amnt = int(user_data[3])
    if with_amnt<= prv_amnt:
        prv_amnt-= with_amnt

    withdraw_here(lineno, user_data, prv_amnt)
    print("Cash Withdrawal Successful :)")
    

def withdraw_here(lineNo, user_found, amnt):
    name = user_found[0] + ','
    uname = user_found[1] + ','
    pass1 = user_found[2] + ','
    pysa = str(amnt) + '\n'
    isSkipped = False
    index = 1
    with open("accounts.txt", 'r') as readObj, open("temp_file", 'w') as writeObj:
        for line in readObj:
            if index != lineNo:
                writeObj.write(line)
            else:
                isSkipped = True
            index += 1
        writeObj.write(name)
        writeObj.write(uname)
        writeObj.write(pass1)
        writeObj.write(pysa)
    if isSkipped:
        os.remove("accounts.txt")
        os.rename("temp_file", "accounts.txt")
    else:
        os.remove("temp_file")
```

```python
def Transfer_money():
    try:
        with open('accounts.txt', 'r') as file:
            file_reader = csv.reader(file)
            cash_t = withdrawal(file_reader)  # cash_t[user_data_t, rem_amnt_t, trans_amnt_t, lineno_t]
            file.seek(0)
            cash_r = receiver(file_reader, cash_t[2])  # cash_r[user_data_r, total_amnt, lineno_r]
            file.seek(0)
            lines_setter(cash_t, cash_r)
            print("Transaction completed Successfully :)\n")
            file.close()
    except:
        print("No File Found!! Make a new account first")


def lines_setter(user_t, user_r):
    name_t = user_t[0][0] + ','
    uname_t = user_t[0][1] + ','
    pass1_t = user_t[0][2] + ','
    pysa_t = str(user_t[1]) + '\n'

    name_r = user_r[0][0] + ','
    uname_r = user_r[0][1] + ','
    pass1_r = user_r[0][2] + ','
    pysa_r = str(user_r[1]) + '\n'
    isSkipped = False
    index = 1
    with open("accounts.txt", 'r') as readObj, open("temp_file", 'w') as writeObj:
        for line in readObj:
            if index != user_t[3] and index != user_r[2]:
                writeObj.write(line)
            else:
                isSkipped = True
            index += 1
        writeObj.write(name_t)
        writeObj.write(uname_t)
        writeObj.write(pass1_t)
        writeObj.write(pysa_t)

        writeObj.write(name_r)
        writeObj.write(uname_r)
        writeObj.write(pass1_r)
        writeObj.write(pysa_r)
    if isSkipped:
        os.remove("accounts.txt")
        os.rename("temp_file", "accounts.txt")
        writeObj.close()
    else:
        os.remove("temp_file")


def withdrawal(file):
    user_data_t,line_no_t = line_user_find(file)
    pass_check(user_data_t)
    with_amnt = int(input("Enter amount you want to Transfer: "))
    prv_amnt = int(user_data_t[3])
    if with_amnt <= prv_amnt:
        rem_amnt = prv_amnt - with_amnt
        return user_data_t, rem_amnt, with_amnt, line_no_t
    else:
        print("amount greater than cash available")


def receiver(file, amnt):
    line_no_r = 1
    uname = input("Enter cash receiver username: ")
    for row in file:
        if row[1] == uname:
            print("User Found:", uname)
            user_data_r = [row[0], row[1], row[2], row[3]]
            prv_amnt = int(user_data_r[3])
            new_amnt = prv_amnt + amnt
            return user_data_r, new_amnt, line_no_r
        else:
            line_no_r += 1

```

```python

def Edit():
    try:
        with open('accounts.txt', 'r') as file:
            file_reader = csv.reader(file)
            Editor(file_reader)
            file.close()
    except:
        print("No file Found! Make an account first!!")


def Editor(file):
    line_no = 1
    uname = input("Enter username: ")
    for row in file:
        if row[1] == uname:
            print("User Found:", uname)
            user_found = [row[0], row[1], row[2], row[3]]
            pass_check(user_found)

            new_name = input("Enter new Name: ")
            new_uname = uname_check()
            new_pass = input("Enter new Password: ")
            amunt = user_found[3]

            edit_here(line_no, new_name, new_uname, new_pass, amunt)
            print("Info Updated Successfully\n")
            break
        else:
            line_no += 1
    else:
        print("USer not Found! <-- OR --> Username Incorrect\n ")


def edit_here(lineNo, name2, uname2, pass2, amnt):
    name = name2 + ','
    uname = uname2 + ','
    pass1 = pass2 + ','
    pysa = str(amnt) + '\n'
    isSkipped = False
    index = 1
    with open("accounts.txt", 'r') as readObj, open("temp_file", 'w') as writeObj:
        for line in readObj:
            if index != lineNo:
                writeObj.write(line)
            else:
                isSkipped = True
            index += 1
        writeObj.write(name)
        writeObj.write(uname)
        writeObj.write(pass1)
        writeObj.write(pysa)
    if isSkipped:
        os.remove("accounts.txt")
        os.rename("temp_file", "accounts.txt")
    else:
        os.remove("temp_file")
```

```python
def uname_check():
    try:
        with open('accounts.txt','r') as f:
            file = csv.reader(f)
            s = True
            c = False
            while s:
                uname = input("Enter new username: ")
                f.seek(0)
                for line in file:
                    if uname in line:
                        print("Username taken Already! Try Again")
                        c = False
                        break
                    else:
                        c = True
                if c == True:
                    s = False
                else:
                    s = True
    except:
        uname = input("Enter Username: ")
    return uname
```
